```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
plt.rcParams.update({'figure.figsize': (10, 7), 'figure.dpi': 120})
```

```python
lrt = pd.read_csv('/content/data-penumpang-lrt-di-provinsi-dki-jakarta-tahun-2020.csv')
df = pd.DataFrame({'date': ['1/1/2020', '2/1/2020', '3/1/2020', '4/1/2020', '5/1/2020', '6/1/2020', '7/1/2020', '8/1/2020', '9/1/2020', '10/1/2020', '11/1/2020', '12/1/2020']})
df['date'] = pd.to_datetime(df['date'])
lrt['tahun'] = df['date']
lrt.drop(['bulan', 'jenis'], axis=1, inplace=True)
lrt['difference_amount'] = lrt['jumlah'].diff()
lrt['difference_pct'] = lrt['jumlah'].pct_change() * 100
lrt.fillna(0, inplace=True)
lrt
```

```python
from fbprophet import Prophet

m = Prophet(interval_width=0.95, 
            changepoint_prior_scale=0.5,
            changepoint_range=0.95, growth='logistic')

m.add_country_holidays(country_name='ID')
m.fit(train)
```

```python
future = m.make_future_dataframe(periods=90)
future
```

```python
future['floor'] = 0
future['cap'] = 180000
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
from fbprophet import Prophet

mrt = pd.read_csv('/content/data-penumpang-mrt-di-provinsi-dki-jakarta-tahun-2020.csv')

df = pd.DataFrame({'date': ['1/1/2020', '2/1/2020', '3/1/2020', '4/1/2020', '5/1/2020', '6/1/2020', '7/1/2020', '8/1/2020', '9/1/2020', '10/1/2020', '11/1/2020', '12/1/2020']})
df['date'] = pd.to_datetime(df['date'])

train = {'ds': df['date'], 'y': mrt['jumlah']}
train = pd.DataFrame(train)
train['ds'] = pd.to_datetime(train['ds'])
train['floor'] = 0
train['cap'] = 4000000

m = Prophet(interval_width=0.95, 
            changepoint_prior_scale=0.5,
            changepoint_range=0.95, growth='logistic')

m.add_country_holidays(country_name='ID')
m.fit(train)

future = m.make_future_dataframe(periods=90)
future['floor'] = 0
future['cap'] = 4000000
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
confirmed_forecast_plot = m.plot(forecast)
```
