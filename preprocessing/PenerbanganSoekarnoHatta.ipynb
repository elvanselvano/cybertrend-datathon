```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette('rainbow')
```

```python
df2020 = pd.read_csv('/content/PenerbanganJKT - 2020.csv')
df2021 = pd.read_csv('/content/PenerbanganJKT - 2021.csv')
```

```python
df2020['year'] = 2020
df2021['year'] = 2021
```

# 2020

```python
df2020 = df2020.rename({'Unnamed: 0' : 'Bulan'}, axis = 1)
```

```python
df2020
```

```python
plt.figure(figsize = (10, 6))
sns.lineplot(x = df2020['Bulan'], y = df2020['Domestik'])
plt.title('Jumlah Penerbangan Domestik di Bandara Soekarno Hatta 2020')
plt.show()
```

Bisa dilihat bahwa situasi dki jakarta sudah mulai membaik mulai pada bulan Mei 2020

```python
plt.figure(figsize = (10, 6))
sns.lineplot(x = df2020['Bulan'], y = df2020['Internasional'])
plt.title('Jumlah Penerbangan Internasional di Bandara Soekarno Hatta 2020')
plt.show()
```

Situasi penerbangan di bandara Soekarno Hatta masih belum kunjung mendapat perkembangan yang berarti

#2021

```python
df2021 = df2021.rename({'Unnamed: 0' : 'Bulan'}, axis = 1)
```

```python
plt.figure(figsize = (10, 6))
sns.lineplot(x = df2021['Bulan'], y = df2021['Domestik'])
plt.title('Jumlah Penerbangan Domestik di Bandara Soekarno Hatta 2021')
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.lineplot(x = df2021['Bulan'], y = df2021['Internasional'])
plt.title('Jumlah Penerbangan Internasional di Bandara Soekarno Hatta 2021')
plt.show()
```

```python
df2 = pd.concat([df2020, df2021])
df2 = df2.reset_index()
df2
```

```python
plt.figure(figsize = (16, 10))
sns.lineplot(data = df2, x = 'Bulan', y = 'Domestik', hue ='year')
plt.ylabel('Jumlah Penumpang')
plt.title('Jumlah Penumpang Domestik Pesawat Udara di Bandara Soekarno Hatta')
plt.show()
```

```python
plt.figure(figsize = (16, 10))
sns.lineplot(data = df2, x = 'Bulan', y = 'Internasional', hue ='year')
plt.ylabel('Jumlah Penumpang')
plt.title('Jumlah Penumpang Internasional Pesawat Udara di Bandara Soekarno Hatta')
plt.show()
```
