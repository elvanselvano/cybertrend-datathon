## 2019

```python
import pandas as pd

df = pd.read_excel('/content/Tingkat Penghunian Kamar pada Hotel Bintang 2019.xlsx', skiprows=2, skipfooter=2)
df = df.dropna().drop('Tahunan', axis=1)
column_names = df.columns.to_list()
column_names[0] = 'name'
df.columns = column_names
df = df.loc[:33]
df['name'] = df['name'].replace({'KEP. BANGKA BELITUNG': 'KEPULAUAN BANGKA BELITUNG	', 'KEP. RIAU': 'KEPULAUAN RIAU	'})
df['year_avg'] = df[column_names[1:]].mean(axis=1)

long_lat = pd.read_csv('https://raw.githubusercontent.com/benangmerah/wilayah/master/datasources/daftar-nama-daerah.csv')
long_lat = long_lat.loc[:32]
long_lat['name'] = long_lat['name'].apply(lambda x: ' '.join(x.split(' ')[1:])).str.upper()
long_lat = long_lat[['name', 'latitude', 'longitude']]

df_master = df.merge(long_lat, on='name')
df_master
```

```python
df_master.to_csv('hotel_2019.csv', index=False)
```
