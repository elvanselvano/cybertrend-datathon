```python
import pandas as pd
import os, re
```

```python
files = [i for i in os.listdir('.') if i.endswith('xlsx')]
```

```python
def process(file):
  year = re.findall('\d+', file)[0]
  df = pd.read_excel(file, skiprows=2)
  cond = df.iloc[:,0] == 'DKI JAKARTA'
  df = df[cond].T.iloc[1:]
  df.columns = ['penghunian kamar']
  df = df[df['penghunian kamar'] != '-']
  df = df.reset_index()
  df['tanggal'] = df['index'] + '-' + year
  df = df.drop('index', axis=1)
  return df[['tanggal', 'penghunian kamar']]
```

```python
dfs = [process(i) for i in files]
df = pd.concat(dfs)
df
```

```python
df.to_csv('tingkat_penghunian_hotel_bintang_jakarta.csv', index=False)
```
