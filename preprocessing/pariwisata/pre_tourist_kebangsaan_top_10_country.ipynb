```python
import pandas as pd
```

```python
base_path = '/content/Kunjungan Wisatawan Mancanegara per bulan Menurut Kebangsaan '
years = [2017, 2018, 2019, 2020, 2021]
```

```python
def process(path, year):
  df = pd.read_excel(path, skiprows=2)
  df.dropna(inplace=True)
  df = df.replace({'-':0})
  col = df.columns.to_list()
  col[0] = 'Kebangsaan'
  df.columns = col
  cond = df['Kebangsaan'].str.match(r'.*(TOTAL)|(A S E A N).*')
  df = df[~cond]
  df['Tahunan'] = df.iloc[:,1:-1].astype(int).sum(axis=1, skipna=True, numeric_only=True)
  df['Tahun'] = year
  return df[['Tahun', 'Kebangsaan', 'Tahunan']].sort_values('Tahunan', ascending=False).head(10)
```

```python
dfs = [process(f'{base_path}{i}.xlsx', i) for i in years]
```

```python
df = pd.concat(dfs)
df
```

```python
df.to_csv('top_10_tourist_kebangsaan_concat.csv')
```
