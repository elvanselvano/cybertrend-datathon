```python
import pandas as pd

df = pd.read_csv('/content/data-jumlah-vaksinasi-covid-19-tahap-3-tiga-menurut-kelurahan-di-provinsi-dki-jakarta-tahun-bulan-juli-2021.csv')
df
```
