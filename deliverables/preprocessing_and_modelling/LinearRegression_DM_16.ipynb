```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/tingkat_penghunian_hotel_bintang_jakarta.csv')
df['month'] = df['tanggal'].str.split('-').str.get(0)
df['month'] = df['month'].str[:3]
df['month'] = np.where(df['month'] == 'Mei', 'May', df['month'])
df['month'] = np.where(df['month'] == 'Agu', 'Aug', df['month'])
df['month'] = np.where(df['month'] == 'Okt', 'Oct', df['month'])
df['month'] = np.where(df['month'] == 'Des', 'Dec', df['month'])
df['year'] = df['tanggal'].str.split('-').str.get(1)
df['date'] = df['month'] + '-' + df['year']
df['date'] = pd.to_datetime(df['date'])
df.drop(['tanggal', 'month', 'year'], axis=1, inplace=True)
df
```

```python
df.to_csv('tingkat_penghunian_hotel_bintang_jakarta.csv', index=False)
```
